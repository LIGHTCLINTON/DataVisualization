```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd

# Generate Data
np.random.seed(0)
cities = ['Nairobi', 'Nakuru', 'Kisumu', 'Mombasa', 'Eldoret']
years = np.arange(2015, 2025)
average_temperatures = np.random.uniform(18, 30, (10, 5))
temperature_df = pd.DataFrame(average_temperatures, columns=cities, index=years)

# Prepare Data for Plotting
average_temps = temperature_df.mean().reset_index()
average_temps.columns = ['City', 'Average Temperature']

# Plotting Bar Chart
plt.figure(figsize=(12, 6))
sns.barplot(data=average_temps, x='City', y='Average Temperature', hue='City', dodge=False, palette='Set2', legend=False)
plt.title('Average Temperatures of Cities Over the Past 10 Years', fontsize=14)
plt.xlabel('Cities', fontsize=12)
plt.ylabel('Average Temperature (°C)', fontsize=12)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd

# Sample data generation
np.random.seed(42)
study_hours = np.random.uniform(1, 10, 50)  # Random study hours between 1 and 10
grades = 50 + study_hours * 5 + np.random.normal(0, 5, 50)  # Linear relation with some noise

# Create a DataFrame
data = pd.DataFrame({'Study Hours': study_hours, 'Grades': grades})

# Plotting the scatter plot with regression line
plt.figure(figsize=(10, 6))
sns.set_theme(style="whitegrid")

# Scatter plot
sns.scatterplot(
    x='Study Hours',
    y='Grades',
    data=data,
    hue='Grades',
    palette='viridis',
    s=100
)

# Regression line
sns.regplot(
    x='Study Hours',
    y='Grades',
    data=data,
    scatter=False,  # Hide scatter points to avoid overlap
    color='red',     # Regression line color
    line_kws={'label': 'Regression Line', 'linewidth': 2}
)

# Adding titles and labels
plt.title('Correlation between Study Hours and Student Grades', fontsize=14)
plt.xlabel('Study Hours', fontsize=12)
plt.ylabel('Grades', fontsize=12)
plt.legend(title='Grades', loc='upper left')
plt.show()
```

```python
import yfinance as yf
import pandas as pd
import plotly.express as px

# Define the ticker symbol (e.g., 'AAPL' for Apple)
ticker_symbol = 'AAPL'

# Fetch data for the past six months
stock_data = yf.download(ticker_symbol, period='6mo')

# Reset index to use 'Date' as a column
stock_data.reset_index(inplace=True)

# Ensure 'Close' is a 1D Series by squeezing the column
# This converts it from a DataFrame column to a pandas Series
fig = px.line(
    stock_data,
    x='Date',
    y=stock_data['Close'].squeeze(),  # Using squeeze() to ensure 1D
    title=f'<b>{ticker_symbol} Stock Prices Over the Past 6 Months</b>',
    labels={'Close': 'Stock Price (USD)', 'Date': 'Date'},
    markers=True
)

# Customize hover information
fig.update_traces(
    mode='lines+markers',
    hovertemplate='<b>Date:</b> %{x}<br><b>Price:</b> $%{y:.2f}'
)

# Update layout for better visualization
fig.update_layout(
    title_x=0.5,  # Centers the title
    xaxis_title='<b>Date</b>',
    yaxis_title='<b>Stock Price (USD)</b>',
    hovermode='x unified',  # Ensures hover applies to all points on the same x-axis
    template='plotly_white',  # Clean look
    xaxis=dict(showgrid=True, showline=True, linewidth=1, linecolor='black'),
    yaxis=dict(showgrid=True, showline=True, linewidth=1, linecolor='black')
)

# Display the plot
fig.show()

```
